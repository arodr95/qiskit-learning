# Binary Numbers - Challenge #1
## Goals
- Initialize a Quantum Citcuit with four qubits
- Set the qubits into the basis state $\ket{1101}$
- Plot the state using the Qiskit QSphere

```python
from qiskit import *
from qiskit_aer import * 
from qiskit.visualization import plot_histogram, array_to_latex, plot_state_qsphere
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(4)
circuit.x([0, 2, 3])
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
plot_state_qsphere(statevector);
```

# Binary Age - Challenge #2
## Goals
- Encode your binary age in qubits
-  I am 25 so I need to set the qubits into the basis state $\ket{11001}$
- Plot the state using the Qiskit QSphere

```python
circuit = QuantumCircuit(5)
circuit.x([0, 3, 4])
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
plot_state_qsphere(statevector);
```

```python
circuit = QuantumCircuit(5)
circuit.y([0, 3, 4])
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
plot_state_qsphere(statevector);
```

Looks like we got the same answer!
<br>
I thought that the color of the vector might be different because of the phase shift, but after thinking about it some this is my conclusion...
<br>
Altough we did use a phase shift and bit flip to get to our result, when a qubit is in it's basis state, the phase shift only represents HOW we got to the other basis state in the qubit (from $\ket{0}$ to $\ket{1}$).
<br>
We don't see any change in the color of the phase because our qubit was never in a quantum state.
