# Controlled NOT (CNOT or CX) Gate

The gates that we have used so far only operate on one qubit at a time.
<br> 
To build a quantum program we also need to create interactions and dependencies in between qubits. 

The CNOT gate conditionally flips the target qubit if the control qubit is $\ket{1}$ by performing a X gate.
<br>
If the control qubit is $\ket{1}$ the target qubit will always have an X gate applied to it. 

$
CNOT =
\begin{bmatrix}
1&0&0&0\cr
0&1&0&0\cr
0&0&0&1\cr
0&0&1&0
\end{bmatrix}
$

Where the 2nd qubit is our control and the 1st qubit is the target...
<br>
<br>
$
CNOT\ket{00}=
\begin{bmatrix}
1&0&0&0\cr
0&1&0&0\cr
0&0&0&1\cr
0&0&1&0
\end{bmatrix}
\begin{bmatrix}
1\cr 0\cr 0\cr 0
\end{bmatrix}
=\begin{bmatrix}
1\cr 0\cr 0\cr 0
\end{bmatrix}
=\ket{00}
$
<br>
<br>
$
CNOT\ket{01}=
\begin{bmatrix}
1&0&0&0\cr
0&1&0&0\cr
0&0&0&1\cr
0&0&1&0
\end{bmatrix}
\begin{bmatrix}
0\cr 1\cr 0\cr 0
\end{bmatrix}
=\begin{bmatrix}
0\cr 1\cr 0\cr 0
\end{bmatrix}
=\ket{01}
$
<br>
<br>
$
CNOT\ket{10}=
\begin{bmatrix}
1&0&0&0\cr
0&1&0&0\cr
0&0&0&1\cr
0&0&1&0
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 1\cr 0
\end{bmatrix}
=\begin{bmatrix}
0\cr 0\cr 0\cr 1
\end{bmatrix}
=\ket{11}
$
<br>
<br>
$
CNOT\ket{11}=
\begin{bmatrix}
1&0&0&0\cr
0&1&0&0\cr
0&0&0&1\cr
0&0&1&0
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 0\cr 1
\end{bmatrix}
=\begin{bmatrix}
0\cr 0\cr 1\cr 0
\end{bmatrix}
=\ket{10}
$

## CX Gate in  Qiskit

### $CX\ket{00}$

```python
from qiskit import *
from qiskit_aer import * 
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(2)
circuit.cx(1, 0)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector}=")
```

```python
plot_bloch_multivector(statevector);
```

### CX on 3 qubits

```python
circuit = QuantumCircuit(3)
circuit.x(0)
circuit.cx(0, 1)
circuit.cx(1, 2)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector}=")
```

```python
plot_bloch_multivector(statevector);
```

What if we switch the order of the CX gates in this example?

```python
circuit = QuantumCircuit(3)
circuit.x(0)
circuit.cx(1, 2)
circuit.cx(0, 1)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector}=")
```

```python
plot_bloch_multivector(statevector);
```

The third qubit does not have a bit switch because the second qubit is $\ket{0}$ when the CX gate with the second qubit as control and third qubit as target is applied. 
<br>
Then we applied another CX gate with control as first qubit and target as second. 
<br>
The first qubit's state was $\ket{1}$ so the second qubit's state is flipped to $\ket{1}$.
<br>
This gives us our final state of $\ket{011}$.
