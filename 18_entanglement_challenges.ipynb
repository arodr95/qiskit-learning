# Entanglement Challenges

## GHZ Challenge #1
Goals:
- Build a quantum circuit that puts three qubits into the $\ket{GHZ}$ state

## Greenberger-Horne-Zeilinger (GHZ) State
A specific type of entangled quantum state with at least three qubits.

$
\ket{GHZ}
=\frac{\ket{000}+\ket{111}}{\sqrt{2}}
=\frac{1}{\sqrt{2}}
\begin{bmatrix}
1\cr 0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 1\cr
\end{bmatrix}
$

### Attempt

```python
from qiskit import * 
from qiskit_aer import *
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.h(0)
circuit.cx(0, 1)
circuit.cx(0, 2)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector} =")
```

```python
plot_bloch_multivector(statevector);
```

```python
circuit.measure_all()
simulator = AerSimulator()
compiled_circuit = transpile(circuit, simulator)
sim_result = simulator.run(compiled_circuit).result()
plot_histogram(sim_result.get_counts())
print(sim_result.get_counts())
```
