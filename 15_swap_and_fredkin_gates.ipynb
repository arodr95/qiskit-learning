# SWAP Gate

Swaps the quantum state of two qubits.
<br>
When the state of the 2 qubits is the same, there is no change to the state.
<br>
When the state is different, the states of the 2 qubits are swapped.
<br>
An important note is that the qubits themselves do not physically trade places, only both of their states change to what the other qubit's was.
<br>
<br>
$
\text{SWAP}
=\begin{bmatrix}
1&0&0&0\cr
0&0&1&0\cr
0&1&0&0\cr
0&0&0&1\cr
\end{bmatrix} 
$

In both these examples, the state of the qubit remains the same because they had the same state as the other qubit!
<br>
<br>
$
\text{SWAP}\ket{00}
=\begin{bmatrix}
1&0&0&0\cr
0&0&1&0\cr
0&1&0&0\cr
0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
1\cr 0\cr 0\cr 0\cr
\end{bmatrix}
=\begin{bmatrix}
1\cr 0\cr 0\cr 0\cr
\end{bmatrix}
=\ket{00}
$
<br>
<br>
$
\text{SWAP}\ket{11}
=\begin{bmatrix}
1&0&0&0\cr
0&0&1&0\cr
0&1&0&0\cr
0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 0\cr 1\cr
\end{bmatrix}
=\begin{bmatrix}
0\cr 0\cr 0\cr 1\cr
\end{bmatrix}
=\ket{11}
$

Let's see what happens in the other possible states of our 2 qubits.
<br>
<br>

$
\text{SWAP}\ket{01}
=\begin{bmatrix}
1&0&0&0\cr
0&0&1&0\cr
0&1&0&0\cr
0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
0\cr 1\cr 0\cr 0\cr
\end{bmatrix}
=\begin{bmatrix}
0\cr 0\cr 1\cr 0\cr
\end{bmatrix}
=\ket{10}
$
<br>
<br>
$
\text{SWAP}\ket{10}
=\begin{bmatrix}
1&0&0&0\cr
0&0&1&0\cr
0&1&0&0\cr
0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 1\cr 0\cr
\end{bmatrix}
=\begin{bmatrix}
0\cr 1\cr 0\cr 0\cr
\end{bmatrix}
=\ket{01}
$

# Fredkin (Controlled-SWAP) Gate
Conditionally swaps two target qubits if a control qubit is $\ket{1}$.
<br>
<br>
$
\text{CSWAP}
=\begin{bmatrix}
1&0&0&0&0&0&0&0\cr
0&1&0&0&0&0&0&0\cr
0&0&1&0&0&0&0&0\cr
0&0&0&1&0&0&0&0\cr
0&0&0&0&1&0&0&0\cr
0&0&0&0&0&0&1&0\cr
0&0&0&0&0&1&0&0\cr
0&0&0&0&0&0&0&1\cr
\end{bmatrix}
$

Let's use the CSWAP gate on a 3-qubit system where the third qubit is our control and the first and second qubit are the targets.

$
\text{CSWAP}\ket{000}
=\begin{bmatrix}
1&0&0&0&0&0&0&0\cr
0&1&0&0&0&0&0&0\cr
0&0&1&0&0&0&0&0\cr
0&0&0&1&0&0&0&0\cr
0&0&0&0&1&0&0&0\cr
0&0&0&0&0&0&1&0\cr
0&0&0&0&0&1&0&0\cr
0&0&0&0&0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
1\cr 0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 0\cr
\end{bmatrix}
= \begin{bmatrix}
1\cr 0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 0\cr
\end{bmatrix}
=\ket{000}
$

$
\text{CSWAP}\ket{100}
=\begin{bmatrix}
1&0&0&0&0&0&0&0\cr
0&1&0&0&0&0&0&0\cr
0&0&1&0&0&0&0&0\cr
0&0&0&1&0&0&0&0\cr
0&0&0&0&1&0&0&0\cr
0&0&0&0&0&0&1&0\cr
0&0&0&0&0&1&0&0\cr
0&0&0&0&0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 0\cr 1\cr 0\cr 0\cr 0\cr 0\cr
\end{bmatrix}
= \begin{bmatrix}
0\cr 0\cr 0\cr 1\cr 0\cr 0\cr 0\cr 0\cr
\end{bmatrix}
=\ket{100}
$

$
\text{CSWAP}\ket{111}
=\begin{bmatrix}
1&0&0&0&0&0&0&0\cr
0&1&0&0&0&0&0&0\cr
0&0&1&0&0&0&0&0\cr
0&0&0&1&0&0&0&0\cr
0&0&0&0&1&0&0&0\cr
0&0&0&0&0&0&1&0\cr
0&0&0&0&0&1&0&0\cr
0&0&0&0&0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 1\cr
\end{bmatrix}
= \begin{bmatrix}
0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 1\cr
\end{bmatrix}
=\ket{111}
$

$
\text{CSWAP}\ket{101}
=\begin{bmatrix}
1&0&0&0&0&0&0&0\cr
0&1&0&0&0&0&0&0\cr
0&0&1&0&0&0&0&0\cr
0&0&0&1&0&0&0&0\cr
0&0&0&0&1&0&0&0\cr
0&0&0&0&0&0&1&0\cr
0&0&0&0&0&1&0&0\cr
0&0&0&0&0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 0\cr 0\cr 0\cr 1\cr 0\cr 0\cr
\end{bmatrix}
= \begin{bmatrix}
0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 1\cr 0\cr
\end{bmatrix}
=\ket{110}
$

$
\text{CSWAP}\ket{110}
=\begin{bmatrix}
1&0&0&0&0&0&0&0\cr
0&1&0&0&0&0&0&0\cr
0&0&1&0&0&0&0&0\cr
0&0&0&1&0&0&0&0\cr
0&0&0&0&1&0&0&0\cr
0&0&0&0&0&0&1&0\cr
0&0&0&0&0&1&0&0\cr
0&0&0&0&0&0&0&1\cr
\end{bmatrix}
\begin{bmatrix}
0\cr 0\cr 0\cr 0\cr 0\cr 0\cr 1\cr 0\cr
\end{bmatrix}
= \begin{bmatrix}
0\cr 0\cr 0\cr 0\cr 0\cr 1\cr 0\cr 0\cr
\end{bmatrix}
=\ket{101}
$

## SWAP and CSWAP in Qiskit

### SWAP in 3 Qubit System

```python
from qiskit import *
from qiskit_aer import * 
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.h(1)
circuit.rx(-pi/4, 2)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector}=")
```

```python
plot_bloch_multivector(statevector);
```

Let's insert a swap gate between the last 2 qubits.

```python
circuit = QuantumCircuit(3)
circuit.h(1)
circuit.rx(-pi/4, 2)
circuit.swap(1, 2)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector}=")
```

```python
plot_bloch_multivector(statevector);
```

### CSWAP in 3 Qubit System

```python
circuit = QuantumCircuit(3)
circuit.h(1)
circuit.rx(-pi/4, 2)
circuit.cswap(0, 1, 2)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector}=")
```

```python
plot_bloch_multivector(statevector);
```

The states of the second and third qubits were not swapped because the first qubit is still in $\ket{0}$.

```python
circuit = QuantumCircuit(3)
circuit.x(0)
circuit.h(1)
circuit.rx(-pi/4, 2)
circuit.cswap(0, 1, 2)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector}=")
```

```python
plot_bloch_multivector(statevector);
```

Now that our first qubit has the X gate and is in state $\ket{1}$ the states of the second and third qubits are swapped.
