# Hadamard Gate 
Transitions our qubits from the classical realm into the quantum!
<br>
<br>
Rotation of $\pi$ around Bloch sphere vector $\begin{bmatrix}1\cr 0\cr 1\end{bmatrix}$,
<br>
<br>
which is the vector pointing halfway between the x-axis and the z-axis.

$
H =
\begin{bmatrix}\frac{1}{\sqrt{2}}&\frac{1}{\sqrt{2}}\cr\frac{1}{\sqrt{2}}&-\frac{1}{\sqrt{2}}\end{bmatrix}
= \frac{1}{\sqrt{2}}\begin{bmatrix}1&1\cr 1&-1\end{bmatrix}
$

## Hadamard Gate on $\ket{0}$

$
H\ket{0} 
=\frac{1}{\sqrt{2}}\begin{bmatrix}1&1\cr 1&-1\end{bmatrix}\begin{bmatrix}1\cr 0\end{bmatrix}
=\begin{bmatrix}\frac{1}{\sqrt{2}}\cr\frac{1}{\sqrt{2}}\end{bmatrix}
=\frac{\ket{0}+\ket{1}}{\sqrt{2}}
=\ket{+}
$
<br>
<br>
$
P(\ket{+}=\ket{0})=|\frac{1}{\sqrt{2}}|^2=\frac{1}{2}
$
<br>
<br>
$
P(\ket{+}=\ket{1})=|\frac{1}{\sqrt{2}}|^2=\frac{1}{2}
$

```python
from qiskit import *
from qiskit_aer import * 
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(1, 1)
circuit.h(0)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
state_vector = sim_result.get_statevector()
array_to_latex(state_vector, prefix="\\text{state\\_vector = }\n")
```

```python
plot_bloch_multivector(state_vector);
```

```python
circuit.measure(0, 0)
circuit.draw(output='mpl');
```

```python
simulator = AerSimulator()
compiled_circuit = transpile(circuit, simulator)
sim_result = simulator.run(compiled_circuit).result()
plot_histogram(sim_result.get_counts())
print(sim_result.get_counts())
```

## Hadamard Gate on $\ket{1}$

$
H\ket{1} 
=\frac{1}{\sqrt{2}}\begin{bmatrix}1&1\cr 1&-1\end{bmatrix}\begin{bmatrix}0\cr 1\end{bmatrix}
=\begin{bmatrix}\frac{1}{\sqrt{2}}\cr-\frac{1}{\sqrt{2}}\end{bmatrix}
=\frac{\ket{0}-\ket{1}}{\sqrt{2}}
=\ket{-}
$
<br>
<br>
$
P(\ket{-}=\ket{0})=|\frac{1}{\sqrt{2}}|^2=\frac{1}{2}
$
<br>
<br>
$
P(\ket{-}=\ket{1})=|-\frac{1}{\sqrt{2}}|^2=\frac{1}{2}
$

```python
circuit = QuantumCircuit(1, 1)
circuit.x(0)
circuit.h(0)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
state_vector = sim_result.get_statevector()
array_to_latex(state_vector, prefix="\\text{state\\_vector = }\n")
```

```python
plot_bloch_multivector(state_vector);
```

```python
circuit.measure(0, 0)
circuit.draw(output='mpl');
```

```python
simulator = AerSimulator()
compiled_circuit = transpile(circuit, simulator)
sim_result = simulator.run(compiled_circuit).result()
plot_histogram(sim_result.get_counts())
print(sim_result.get_counts())
```
