# Phase Shift Gates - $S, S^{\dagger}, T, T^{\dagger}$

## $S$ Gate

Rotation of $\frac{\pi}{2}$ around the z-axis of the Bloch sphere.

$
S
=\begin{bmatrix}1&0\cr 0&i\end{bmatrix}
=\begin{bmatrix}1&0\cr 0&e^{\frac{i\pi}{2}}\end{bmatrix}
$

$
S\ket{+}
=\begin{bmatrix}1&0\cr 0&i\end{bmatrix}\begin{bmatrix}\frac{1}{\sqrt{2}}\cr\frac{1}{\sqrt{2}}\end{bmatrix}
=\begin{bmatrix}\frac{1}{\sqrt{2}}\cr\frac{i}{\sqrt{2}}\end{bmatrix}
=\ket{i}
$
<br>
<br>
$
S\ket{i}
=\begin{bmatrix}1&0\cr 0&i\end{bmatrix}\begin{bmatrix}\frac{1}{\sqrt{2}}\cr\frac{i}{\sqrt{2}}\end{bmatrix}
=\begin{bmatrix}\frac{1}{\sqrt{2}}\cr\frac{-1}{\sqrt{2}}\end{bmatrix}
=\ket{-}
$
<br>
<br>
Unlike the previous gates, the S gate only rotates the vector by $\frac{\pi}{2}$ around the z-axis so it is not the inverse of itself!

$
SS
=\begin{bmatrix}1&0\cr 0&i\end{bmatrix}\begin{bmatrix}1&0\cr 0&i\end{bmatrix}
=\begin{bmatrix}1&0\cr 0&-1\end{bmatrix}
= Z
$
<br>
<br>
Multiplying the S matrix by itself will give us the equivalent of the Z gate.

Well what if we want to rotate around the y-axis by $\frac{3\pi}{2}$?
<br>
Should we apply three S gates?
<br>
While technically equivalent, instead of using three gates to accomplish this we can use just one with the S-Dagger gate.

## $S^{\dagger}$ Gate

Rotation of $\frac{-\pi}{2}$ around the z-axis of the Bloch sphere.

$
S^{\dagger}
=\begin{bmatrix}1&0\cr 0&-i\end{bmatrix}
=\begin{bmatrix}1&0\cr 0&e^{\frac{-i\pi}{2}}\end{bmatrix}
$

This is the conjugate transpose of the regular S gate matrix.

$
SS^{\dagger}
=\begin{bmatrix}1&0\cr 0&i\end{bmatrix}\begin{bmatrix}1&0\cr 0&-i\end{bmatrix}
=\begin{bmatrix}1&0\cr 0&1\end{bmatrix}
=I
$

If we multiply the S gate matrix by the S-Dagger gate matrix, we get the identity matrix, $I$.
<br>
This is a property of matrices, any matrix multiplied with it's conjugate transpose will give the identity matrix.
<br>
The identity matrix will not change the state of our qubit, so when we apply the S gate and then the S-Dagger gate, our qubit will return to it's initial state.

## $T$ Gate

Rotation of $\frac{\pi}{4}$ around the z-axis of the Bloch sphere.

$
T
=\begin{bmatrix}1&0\cr 0&e^{\frac{i\pi}{4}}\end{bmatrix}
$

## $T^{\dagger}$ Gate

Rotation of $\frac{-\pi}{4}$ around the z-axis of the Bloch sphere.

$
T
=\begin{bmatrix}1&0\cr 0&e^{\frac{-i\pi}{4}}\end{bmatrix}
$

## $S$ and $S^\dagger$ Gates in Qiskit

```python
from qiskit import *
from qiskit_aer import * 
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.h([0, 1, 2])
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector = }\n")
```

```python
plot_bloch_multivector(statevector);
```

```python
# Apply a S gate to the middle qubit and a S-Dagger gate to the last qubit 
circuit.s(1)
circuit.sdg(2)
circuit.draw(output='mpl');
```

```python
simulator = StatevectorSimulator()
sim_result = simulator.run(circuit).result()
statevector = sim_result.get_statevector()
array_to_latex(statevector, prefix="\\text{statevector = }\n")
```

```python
plot_bloch_multivector(statevector);
```
